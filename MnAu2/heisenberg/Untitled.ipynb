```python
data = pd.read_csv("snapshots/config_1000.txt", sep="\t", header=None)
data.columns = ["n", "theta", "phi", "spinX", "spinY", "spinZ"]
```

```python
xyz = np.zeros([len(data), 3])
xCells = np.array([0.0, 0.5])
yCells = np.array([0.0, 0.5])
zCells = np.array([0.0, 0.5])

while (2*m**3 < len(data)):
    m = m+1

    
n = 0
for i in range(m):
    for j in range(m):
        for k in range(m):
            for p in range(2):
                if (n < len(data) ):
                    xyz[n][0] = i + xCells[p] 
                    xyz[n][1] = j + yCells[p] 
                    xyz[n][2] = k + zCells[p] 
                    n = n+1
            
#positions = pd.concat([pd.DataFrame(xyz[:,0]),pd.DataFrame(xyz[:,1]),pd.DataFrame(xyz[:,2])], axis =1)

data["x"] = xyz[:,0]
data["y"] = xyz[:,1]
data["z"] = xyz[:,2]
data

```

```python
fig = plt.figure()
fig.subplots_adjust(hspace=0.02, wspace=0.02)
fig, ax = plt.subplots()
for i in range(1, 7):
    layer1 = data[data.z == i]
    spinXAve = np.mean(layer1.spinX)
    spinYAve = np.mean(layer1.spinY)
    ax = fig.add_subplot(2, 3, i)
    ax.quiver(layer1.x,layer1.y,spinXAve,spinYAve, alpha = 0.5)
    ax.xaxis.set_ticks([])
    ax.yaxis.set_ticks([])
    ax.set_aspect('equal')


```
