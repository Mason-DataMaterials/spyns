```python
from pathlib import Path
from sys import stdout
import numpy as np
import pandas as pd
pd.set_option("display.colheader_justify", "left")
pd.set_option("display.html.border", 0)
html_table_style = {"selector": "th", "props": [("text-align", "left")]}
```

```python
def pattern_tables( layer, p_type):
    n_row, n_column = layer.shape
    au_layer = pd.DataFrame(np.zeros((2,n_column)))
    patterns = pd.concat([ layer[0:1],au_layer,
                           layer[1:2],au_layer,
                           layer[2:3],au_layer,
                           layer[3:4],au_layer,
                           layer[4:5],au_layer,
                           layer[5:6],au_layer,
                           layer[6:7],au_layer,
                           layer[7:8],au_layer], ignore_index=True)
    
    p_types =[]
    p_types.append("nm")
    index = np.arange(1,len(patterns.columns))
    for i in index:
        p_types.append(p_type+str(i))
    patterns.columns = p_types
    
    return patterns
```

```python
c_layer=pd.read_csv('data/mg_collinear_patterns', header=None)
c_patterns = pattern_tables( c_layer, "c-")
row,col = c_patterns.shape
c_matrix = c_patterns
map_Mn = {1: "Mn", -1:"Fe", 0: "Au"}
c_list = []
for i in range(0,col):
    c_list.append("c-"+str(i))
    
    
for i in range(1,col):
    c_matrix[c_list[i]] = c_matrix[c_list[i]].map(map_Mn)
#c_matrix["nm"] = c_matrix["nm"].map({0:"Mn"})

c_matrix.to_csv("data/pattern_images/collinear.list", header=False, index=False, sep='\t')
```

```python
z_layer=pd.read_csv('data/mg_1x1x4_patterns', header=None)
z_patterns = pattern_tables( z_layer, "z-")

row,col = z_patterns.shape
z_matrix = z_patterns
map_Mn = {1: "Mn", -1:"Fe", 0: "Au"}
z_list = []
for i in range(0,col):
    z_list.append("z-"+str(i))
    
    
for i in range(1,col):
    z_matrix[z_list[i]] = z_matrix[z_list[i]].map(map_Mn)
#z_matrix["nm"] = z_matrix["nm"].map({0:"Mn"})
z_matrix.to_csv("data/pattern_images/1x1x4.list", header=False, index=False, sep='\t')
```
